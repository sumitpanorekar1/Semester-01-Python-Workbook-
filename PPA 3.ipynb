### Question 1
Write a program to find the maximum and minimum element in a tuple without using max() and min().

Here, I am using a loop to compare each element and find maximum and minimum values.

```python
numbers = (10, 5, 20, 3, 15)

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

### Question 2
Write a program to count the occurrence of each element in a tuple.

I am using a loop and a dictionary to keep count of each element.

```python
data = (1, 2, 3, 2, 1, 3, 2)

count_dict = {}

for item in data:
    if item in count_dict:
        count_dict[item] = count_dict[item] + 1
    else:
        count_dict[item] = 1

print("Element counts:")
for key in count_dict:
    print(key, ":", count_dict[key])
```

### Question 3
Write a program to remove duplicates from a tuple.

I am creating a new tuple by checking if elements are already added.

```python
original_tuple = (1, 2, 3, 2, 1, 4, 5)

unique_list = []

for value in original_tuple:
    if value not in unique_list:
        unique_list.append(value)

unique_tuple = tuple(unique_list)

print("Tuple without duplicates:", unique_tuple)
```
