### Question 1
Write a program to count the number of vowels in a given string.

In this program, I am using a loop to check each character and count vowels.

```python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"
vowel_count = 0

for char in text:
    if char in vowels:
        vowel_count = vowel_count + 1

print("Number of vowels:", vowel_count)
```

### Question 2
Write a program to check if a string is a palindrome.

Here, I am reversing the string using a loop and comparing it with the original.

```python
string = input("Enter a string: ")

reversed_string = ""

for char in string:
    reversed_string = char + reversed_string

if string == reversed_string:
    print("The string is a palindrome")
else:
    print("The string is not a palindrome")
```

### Question 3
Write a program to reverse a string without using built-in functions.

```python
original_string = input("Enter a string: ")

reverse_result = ""

for letter in original_string:
    reverse_result = letter + reverse_result

print("Reversed string:", reverse_result)
```
