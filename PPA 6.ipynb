### Question 1
Write a function to find the maximum of three numbers.

*Approach:* I will compare the numbers using if-else conditions inside a function.

```python
def find_maximum(num1, num2, num3):
    # Comparing three numbers step by step
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

result = find_maximum(10, 25, 15)
print('Maximum number is:', result)
```

### Question 2
Check whether a number is prime.

*Approach:* I will use a loop to check if the number is divisible by any number other than 1 and itself.

```python
def check_prime(number):
    if number <= 1:
        return False

    for i in range(2, number):
        if number % i == 0:
            return False
    return True

print(check_prime(7))
```
