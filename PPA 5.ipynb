### Question 1
Create a dictionary of roll numbers and student names.

*Approach:* I will store roll numbers as keys and names as values inside a dictionary.

```python
# Creating a dictionary to store student details
student_details = {
    20997: 'Sumit',
    20998: 'Amit',
    20999: 'Ravi'
}

# Displaying dictionary using a loop
for roll_number in student_details:
    print(roll_number, ':', student_details[roll_number])
```

### Question 2
Count the frequency of each word in a sentence.

*Approach:* I will split the sentence into words and then count each word using a dictionary.

```python
sentence = 'python is easy and python is fun'
word_list = sentence.split()

word_count = {}

# Counting frequency of each word
for word in word_list:
    if word in word_count:
        word_count[word] = word_count[word] + 1
    else:
        word_count[word] = 1

print(word_count)
```
