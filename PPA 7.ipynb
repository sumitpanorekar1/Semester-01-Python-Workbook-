### Question 1
Create a Student class and display student details.

*Approach:* I will create a class with a constructor and a method to display student information.

```python
class Student:
    # Constructor to initialize values
    def __init__(self, student_name, roll_number):
        self.student_name = student_name
        self.roll_number = roll_number

    # Method to display details
    def display_details(self):
        print('Student Name:', self.student_name)
        print('Roll Number:', self.roll_number)

# Creating an object of Student class
student1 = Student('Sumit', 20997)
student1.display_details()
```

### Question 2
Demonstrate inheritance.

*Approach:* I will create a base class and then inherit it into a child class using inheritance.

```python
class Person:
    def __init__(self, name):
        self.name = name

class CollegeStudent(Person):
    def __init__(self, name, roll_number):
        Person.__init__(self, name)
        self.roll_number = roll_number

    def show_details(self):
        print('Name:', self.name)
        print('Roll Number:', self.roll_number)

student = CollegeStudent('Sumit', 20997)
student.show_details()
```
