### Question 1
Write a Python Program to Calculate the Average of Numbers in a Given List.

Here, I am using a list and a loop to add all numbers and then divide by count.

```python
numbers = [10, 20, 30, 40, 50]

total_sum = 0

for num in numbers:
    total_sum = total_sum + num

average = total_sum / len(numbers)

print("Average is:", average)
```

### Question 2
Accept 6 integers and print DUPLICATES if any values repeat.

```python
values = [32, 10, 45, 90, 45, 6]

unique_values = []
duplicate_found = False

for val in values:
    if val in unique_values:
        duplicate_found = True
    else:
        unique_values.append(val)

if duplicate_found:
    print("DUPLICATES")
else:
    print("ALL UNIQUE")
```

### Question 3
Check if number is between 1 and 50.

```python
number = int(input("Enter a number: "))

if number >= 1 and number <= 50:
    print("Ok")
else:
    print("Out of range")
```
